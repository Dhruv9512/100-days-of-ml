##**Index_col parameter**

Use to set column as an defult index.

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv",index_col="longitude")
print(df)

# Longitude become an index here.
```

##**Header parameter**
Use to make the rows as an column.

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv",header=1)
print(df)

#Here row 1 become the column
```

##**usecols parameter**
At the time of import the data you import only those column which you are going to use for tranning the model.

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv",usecols=["longitude","latitude"])
print(df)
```

##**skiprows parameter**
Use to skip the rows during importing the data.

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv",skiprows=lambda a:a%2==0)
print(df)

# Here I skip the rows which index are even.
```

##**parse_dates**
Default date import as string due to which we can not perform the filters.
Therefore we use this parameter to change the date into int data type.

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv",parse_dates=["column name"])
print(df)

# I don't have the other type encoding data so it will give error.
```

##**convertors parameter**
converter use to apply transformer in a column.

```python
def change(n):
   if n<'0':
      return 2
   else:
      return n

import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv",converters={"longitude":change})
print(df)
```

##**na_values parameters**
When there is a value in data which does not make any sense like "$","-"etc then we use this parameter and make them NAN values.

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv",na_values=[15.000000])
print(df)
```
